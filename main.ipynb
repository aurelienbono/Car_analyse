```python
fig ,axe =plt.subplots(2,2)
fig.set_size_inches(20,10)
axe[0,0].sns.countplot(x=df['Year'] , data=df , orient="90" )
axe[0,1].sns.countplot(x=df['Fuel_Type']) 
axe[1,0].sns.countplot(x=df['Seller_Type'] , palette='crest')
axe[1,1].sns.countplot(x=df['Transmission'])
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x=df['Year'] , data=df , orient="90" )
plt.title("Le Nombre de voiture par année ")
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x=df['Fuel_Type']) 
plt.title("Le nombre de voiture en function du type de carburant ") 
```

```python
plt.figure(figsize=(9,5)) 
sns.countplot(x=df['Seller_Type'] , palette='crest')
plt.title("Nombre de voiture en fonction du type de  vendeur ")
plt.show()
```

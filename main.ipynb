```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
df = pd.read_csv("data/KaggleCarData.csv")
```

```python
df.head(20)
```

```python
df.isnull().sum()/len(df)*100
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x=df['Year'] , data=df , orient="90" )
plt.title("Le Nombre de voiture par ann√©e ")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x=df['Fuel_Type']) 
plt.title("Le nombre de voiture en function du type de carburant ") 
plt.show()
```

```python
plt.figure(figsize=(9,5)) 
sns.countplot(x=df['Seller_Type'] , palette='crest')
plt.title("Nombre de voiture en fonction du type de  vendeur ")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x=df['Transmission'])
plt.title("Nombre total de voiture en fonction de le transmission ")
plt.show()
```

```python
df.groupby(['Year'])['Selling_Price'].sum()
```

```python
df.groupby(['Year'])['Present_Price'].sum().plot(figsize=(20,10) , marker="x", linestyle="--")
plt.ylabel("Present_Price")
plt.title("Volume de vente de vehicule en fonction des ann√©e")
plt.show()
```

```python
plt.figure(figsize=(20,10))
df.groupby('Seller_Type')['Kms_Driven'].value_counts(ascending=True).plot()
plt.show()
```

```python
df.groupby('Year')['Present_Price','Selling_Price'].sum().plot.barh(color=sns.color_palette('Set1'), figsize=(20,10))
plt.title("Les annees en fonction present price et du  selleing price ")
plt.xlabel("price")
plt.show()
```
